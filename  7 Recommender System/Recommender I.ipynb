```python
import pandas as pd
import numpy as np
```

```python
colums = ['user_id', 'item_id', 'rating', 'timestamp']


df = pd.read_csv('u.data',sep='\t', names=colums)
```

```python
movie_titles = pd.read_csv("Movie_Id_Titles")
```

```python
df.head()
```

```python
movie_titles.head()
```

```python
df = pd.merge(df,movie_titles,on='item_id')
```

```python
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white') #background white
```

```python
df.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
df.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
ratings.head()
```

```python
ratings['num of users'] = pd.DataFrame(df.groupby('title')['rating'].count())
ratings.head()
```

```python
plt.figure(figsize=(10,4))
ratings['num of users'].hist(bins=70)
```

```python
plt.figure(figsize=(10,4))
ratings['rating'].hist(bins=70)
```

```python
moviemat = df.pivot_table(index='user_id',columns='title',values='rating')
moviemat.head()
```

```python
ratings.sort_values('num of users',ascending=False).head(10)
```

```python
ratings.head()
```

```python
starwars_user_ratings = moviemat['Star Wars (1977)']
liarliar_user_ratings = moviemat['Liar Liar (1997)']
starwars_user_ratings.head()
```

```python
similar_to_starwars = moviemat.corrwith(starwars_user_ratings)
similar_to_liarliar = moviemat.corrwith(liarliar_user_ratings)
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars,columns=['Correlation'])
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```

```python
corr_starwars.sort_values('Correlation',ascending=False).head(10)
```

```python
corr_starwars = corr_starwars.join(ratings['num of users'])
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['num of users']>100].sort_values('Correlation',ascending=False)
```

```python
corr_liarliar = pd.DataFrame(similar_to_liarliar,columns=['Correlation'])
corr_liarliar.dropna(inplace=True)
corr_liarliar = corr_liarliar.join(ratings['num of users'])
corr_liarliar[corr_liarliar['num of users']>100].sort_values('Correlation',ascending=False)
```
